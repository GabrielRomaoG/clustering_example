```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from etl import preprocess_localiza_data
```

```python
data = preprocess_localiza_data(carros_csv="data/carros.csv", vendas_csv="data/vendas.csv")
```

```python
data.describe()
```

```python
categorical_columns = data.select_dtypes('object').columns.to_list()
categorical_columns.remove("modelo")

for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=column, data=data)
    plt.title(f"Quantidade por {column}")
    plt.show()
```

```python
numerical_columns = data.select_dtypes('number').columns

for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=data, x=column)
    plt.title(f"Histograma da {column}")
    plt.show()
```
